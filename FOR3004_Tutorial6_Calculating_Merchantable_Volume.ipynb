```python
%%R

#Read in your groups data

library(readr)

stand_data <- read.csv("Group1_totals.csv")

numrow <- nrow(stand_data)

#Get MISTR functions 
library(MISTR)

#Initialize the empty vector
store_mv <- vector(mode="numeric",length=numrow) 

#Loop through all rows of the totals file for your group
for (row in 1:numrow) {
  
  sp <- toupper(stand_data[row, "code"])
  area_km  <- stand_data[row, "total_area_km2"]
  site_class  <- stand_data[row, "site_class"]
  height  <- stand_data[row, "height"]
  age <- stand_data[row, "age"]
  
  store_mv[row] <- NA
  
  if (!is.nan(height) & !(sp=='BSH') & !(sp=='CW') & !(sp=='ISL') & 
      !(sp=='MR') & !(sp=='OMS')  & !(sp=='PB')  & !(sp=='PT') & !(sp=='TMS')
      & !(sp=='UCL')& !(sp=='WAT')) {
    
    si <- MISTR::calc_SI(H = height, age = age, SppId = sp)
    bha <- MISTR::get_BHA(sp)
    ba <- MISTR::calc_BA(si, bha, sp,  stocking=1, Origin = "N", PlantedSpp = NA) #Origin N means natural not planted
    n <- calc_N(si, bha, ba, sp, stocking=1, Origin = "N", PlantedSpp = NA)
    qmd <- MISTR::calc_QMD(ba, n)
    gtv <- MISTR::calc_GTV(ba, height, sp, Origin = "N", PlantedSpp = NA)
    gmv <- MISTR::calc_GMV(gtv, qmd, sp, Origin = "N", PlantedSpp = NA) #m3/ha
    mv_total <- gmv*area_km*100 #convert km2 to ha by *100

    if (length(mv_total) >0) {
        store_mv[row] <- mv_total}
    else {store_mv[row] <- NA}
    
    
  }
  else {
    store_mv[row] <- NA
  }
}


stand_data <- cbind(stand_data, Merchantable_Volume=store_mv)
print(stand_data)
write.csv(stand_data, "merchantable_vol_Group1.csv", row.names=FALSE)
```
